# PyCity Schools Analysis

```python
import os
import csv
import pandas as pd

schools = pd.read_csv("raw_data/schools_complete.csv")
students = pd.read_csv("raw_data/students_complete.csv")
```

```python
complete = pd.merge(students,schools, left_on='school', right_on='name')
complete = complete.drop(complete.columns[[4]], axis=1)
complete = complete.rename(index=str, columns={'name_x':'name','name_y':'school'})
complete.head()
```

```python
schools.head()
```

## District Summary


```python
numschools = len(schools)
numstudents = len(complete)
sumbudget = sum(schools['budget'])
avemath = complete['math_score'].mean()
averead = complete['reading_score'].mean()
passmath = ((complete['math_score'] >= 70).sum())/numstudents * 100
passread = ((complete['reading_score'] >= 70).sum())/numstudents * 100
avepass = (passmath+passread)/2
District = pd.DataFrame(columns=['Total Schools','Total Students','Total Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing Rate'], 
                        
                        data=[[numschools,numstudents,sumbudget,avemath,averead,passmath,passread,avepass]])
                        
District['Total Budget'] = District['Total Budget'].map('${:,.2f}'.format)
District
```

## School Summary

```python
school = pd.DataFrame(columns=['School_Type', 'Total_Students', 'Total_School_Budget','Per_Student_Budget','Average_Math_Score','Average_Reading_Score','%_Passing_Math','%_Passing_Reading','%_Overall_Passing_Rate'])
school = school.assign(School_Type=complete.groupby(['school'])['type'].unique())
school = school.assign(Total_Students=complete.groupby(['school'])['Student ID'].count())
school = school.assign(Total_School_Budget=complete.groupby(['school'])['budget'].unique())
school = school.assign(Per_Student_Budget=((school['Total_School_Budget'])/(school['Total_Students'])))
school = school.assign(Average_Math_Score=complete.groupby(['school'])['math_score'].mean())
school = school.assign(Average_Reading_Score=complete.groupby(['school'])['reading_score'].mean())

school.columns = [c.replace('_', ' ') for c in school.columns]
school.index.names = ['']
school['School Type']=school['School Type'].str[0]
school['Total School Budget']=school['Total School Budget'].str[0]
school['Per Student Budget']=school['Per Student Budget'].str[0]
school['Total School Budget']=school['Total School Budget'].map('${:,.2f}'.format)
school['Per Student Budget']=school['Per Student Budget'].map('${:,.2f}'.format)
school

```
