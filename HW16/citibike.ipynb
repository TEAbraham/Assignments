```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
total = pd.read_csv('data/201801-citibike-tripdata.csv')
feb = pd.read_csv('data/201802-citibike-tripdata.csv')
mar = pd.read_csv('data/201803-citibike-tripdata.csv')
apr = pd.read_csv('data/201804-citibike-tripdata.csv')
may = pd.read_csv('data/201805-citibike-tripdata.csv')
jun = pd.read_csv('data/201806-citibike-tripdata.csv')
jul = pd.read_csv('data/201807-citibike-tripdata.csv')
aug = pd.read_csv('data/201808-citibike-tripdata.csv')
```

```python
total.append(feb, ignore_index=True)
total.append(mar, ignore_index=True)
total.append(apr, ignore_index=True)
total.append(may, ignore_index=True)
total.append(jun, ignore_index=True)
total.append(jul, ignore_index=True)
total.append(aug, ignore_index=True)
```
